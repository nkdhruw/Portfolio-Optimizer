```python
%matplotlib inline
import numpy as np
import pandas as pd
from datetime import date
from nsepy import get_history
import matplotlib.pyplot as plt
```

```python
tickers = ['CIPLA','IDEA','WONDERLA','PVR','ALKEM']
data = {}
for ticker in tickers:
    data[ticker] = get_history(symbol=ticker, start=date(2001,1,1), end=date(2018,9,30))
```

```python
data['PVR']
```

```python
data['CIPLA'].to_csv('CIPLA.csv')
data['IDEA'].to_csv('IDEA.csv')
data['WONDERLA'].to_csv('WONDERLA.csv')
data['PVR'].to_csv('PVR.csv')
data['ALKEM'].to_csv('ALKEM.csv')
```

```python
data = data[['Close']]
data['P1'] = data['Close'].shift(1)
data.dropna(inplace=True)
```

```python
#wipro['LogReturn'] = 100*np.log(wipro['Close']/wipro['P1'])
data['LogReturn'] = 100*(data['Close']/data['P1']-1)
data.head()
```

```python
plt.hist(data['LogReturn'], bins=20)
```

```python
dailyVol = data['LogReturn'].std()
annVol = dailyVol*np.sqrt(252)
print('Daily volatility:', dailyVol)
print('Annual volatility:', annVol)
```

```python
dailyAvg = data['LogReturn'].mean()
```

```python
tickers = ['ASHOKLEY','MARUTI','TATAMOTORS']
priceVolData = {}
for ticker in tickers:
    priceVolData[ticker]= pd.read_csv('HISTORICAL_DATA/'+ticker+'_data.csv', index_col=0)
```

```python
start_date = '2016-09-01'
end_date = '2017-02-28'
for ticker in tickers:
    priceVolData[ticker] = priceVolData[ticker][(priceVolData[ticker].index>=start_date)&(priceVolData[ticker].index<=end_date)]
```

```python
data = pd.DataFrame()
for i, ticker in enumerate(tickers):
    data['stock'+str(i+1)] = priceVolData[ticker]['close']
stocks = data.columns.values
```

```python
for i, stock in enumerate(stocks):
    data['P'+str(i+1)] = data[stock].shift(1)
data.dropna(inplace=True)
```

```python
data.head()
```

```python
for i, stock in enumerate(stocks):
    data['Return'+str(i+1)] = 100*(data['stock'+str(i+1)]/data['P'+str(i+1)]-1)
```
